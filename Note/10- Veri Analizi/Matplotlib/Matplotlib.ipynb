# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x=np.arange(1,6)
```

```python
x
```

```python
y=np.arange(2,11,2)
```

```python
y
```

```python
plt.plot(x,y,"red")
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel("X Değerleri")
plt.ylabel("Y Değerleri")
plt.title("x=y2 Grafiği")
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,"red")
plt.subplot(2,2,2)
plt.plot(y,x,"blue")
plt.subplot(2,2,3)
plt.plot(y,y,"black")
plt.subplot(2,2,4)
plt.plot(x,y,"green")
plt.show()
```

```python
fig = plt.figure()
axes=fig.add_axes([0,0,1,1])

axes.plot(x,y,"red")
axes.set_xlabel("X Etiketi")
axes.set_ylabel("Y Etiketi")
axes.set_title("X ve Y Grafiği")
plt.show()
```

```python
fig = plt.figure(figsize=(8,4))

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.4,0.5,0.4,0.3])
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
plt.tight_layout()
```

```python
fig = plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,x**2,color="red",lw=9,ls="-.")
```

```python
fig = plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,x**2,color="red",marker="o",markersize=20,markerfacecolor="black")
```
